# Reading data with Open

```python
import pandas as pd
import os

mainpath = "../datasets/"
data = open(mainpath + "customer-churn-model/Customer Churn Model.txt", "r")

cols = data.readline().strip().split(",")
print(cols, '\n')

n_cols = len(cols)
print(n_cols, '\n')


counter = 0
dict = {}

for col in cols:
    dict[col] = []
    
dict 
```

```python
# Reading file - matrix (x,y)
for line in data:
    values = line.strip().split(",")
    for i in range(len(cols)):
        dict[cols[i]].append(values[i])
    counter += 1

print ("Data set: filas - %d, columnas %d"%(counter, n_cols))
```

# Converting object to data frame

```python
dataframe = pd.DataFrame(dict)
dataframe.head()
```

# Customizing a dataframe

```python
infile = mainpath + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "customer-churn-model/Tabs Customer Churn Model.txt"

with open(infile, "r") as inf:
    with open(outfile, "w") as ouf:
        for line in inf:
            fields = line.strip().split(",")
            ouf.write("\t".join(fields))
            ouf.write("\n")

anotherframe = pd.read_csv(outfile, sep = "\t")
anotherframe.head()
```
