# Data Imputation

## Remove missing values

Los valores que faltan en un data set pueden ser por dos razones:
* Extracción de los datos
* Recolección de los datos

```python
data.dropna(axis=0, how="all").head()
```

```python
data.dropna(axis=0, how="any").head()
```

## Fill missing values

### Custom value

```python
data.fillna(0).head()
```

```python
data.fillna('desconocido').head()
```

```python
data['body'] = data['body'].fillna(0)
data.head()
```

```python
data['home.dest'] = data['home.dest'].fillna('desconocido')
data.tail()
```

```python
# Get sum of all null values
pd.isnull(data['age']).values.ravel().sum()

mean = data['age'].mean()
print(mean)

# Filling with the mean
data['age'] = data['age'].fillna(mean)
print(data['age'][1291])
data.tail()
```

### ffill

```python
data_ff = pd.read_csv(fullpath)
print(data_ff['age'][1291])

# valor próximo hacia adelante
data_ff['age'] = data_ff['age'].fillna(method='ffill')
data_ff[1290:1295]
```

### backfill

```python
data_bf = pd.read_csv(fullpath)
print(data_ff['age'][1291])

# valor próximo hacia atrás
data_bf['age'] = data_bf['age'].fillna(method='backfill')
data_ff[1290:1295]
```

## Dummy variables

```python
# Data as flattened array
print(data['sex'].ravel())

dummy_sex = pd.get_dummies(data['sex'], prefix='sex')
dummy_sex.head()
```

## Removing columns

```python
datawithoutsex = data.drop(['sex'], axis = 1)
datawithoutsex.head()
```

## Creating custom dataframe

```python
pd.concat([datawithoutsex, dummy_sex], axis = 1).head()
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix = var_name)
    df = df.drop([var_name], axis = 1)
    df = pd.concat([df, dummy], axis = 1)
    return df

d = pd.read_csv(fullpath)
createDummies(d, 'sex').head()
```
